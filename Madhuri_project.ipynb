```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import scipy as sp
import sklearn as sk
```

```python
df=pd.read_csv('combined_data_BY_METAL_LEVEL.csv')
df.head()
```

Now creating some Exploratory Data Analysis (EDA) for our data to get some useful findings. 

```python
df['Plan Type'].value_counts().plot(kind="pie")
```

This above barplot shows that PPO plan types are most common among counties.

```python
df['Issuer Name'].value_counts().plot(kind='bar')
```

The above bar chart shows the distribution of most used to least used Issuer accross Illinois counties. Out of them, Blue Cross Blue Shield of Illinois seem to be the most commonly used Issuer.

```python
f, ax = plt.subplots(1, 1, figsize=(12, 8))
ax = sns.barplot(x = 'Plan Type', y = 'Median household income', data=df, hue='Issuer Name', palette='Set1')
```

From the above plot, it can be observerd that Plan Type PPO is provided by majority of the Issuers and Issuer Aetna is preferred mostly by people whose Median household income is the highest.

```python
ax=sns.scatterplot(y='Rating Area',x='County Type',data=df)
plt.show()
```

From the above scatterplot, it can be observed with respect to the rating areas assigned to each county tyoe that Illinois has more Urban Counties with All kinds of Rating areas while Rural Counties in Illinois has only selected Rating areas.
